```python
# TEST n°4 RÉGRESSION LINÉAIRE
# Prédire la couverture forestière à partir de la proportion de pixels verts

import numpy as np
import matplotlib.pyplot as plt

print("--RÉGRESSION LINÉAIRE--")

# 1. CRÉER LES DONNÉES D'ENTRAÎNEMENT

# Quelles sont les données : pourcentages de pixels verts(%) → couverture de la foret(%)
# Dans la réalité, ces données viennent d'images annotées

X_train = np.array([10, 20, 30, 40, 50, 60, 70, 80])  # Entrées(pixels verts)
y_train = np.array([5, 15, 25, 38, 48, 58, 70, 82])   # Sorties (couverture réelle)

print("\n Données d'entraînement\n")
print("Nombre d'exemples :", len(X_train))
print("\nPixels verts (%) :", X_train)
print("Couverture réelle (%) :", y_train)

# Visualiser les données
plt.figure(figsize=(10, 6))
plt.scatter(X_train, y_train, color="green", s=100, alpha=0.6, label="Données réelles")
plt.xlabel("Pourcentage de pixels verts (%)", fontsize=12)
plt.ylabel("Couverture forestière (%)", fontsize=12)
plt.title("Données d'entraînement : pixels verts vs couverture forestière", fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

print("\nConclusion : Les points suivent globalement une tendance linéaire")
print(" On essaye de tracer une ligne qui passe au mieux par ces points")

# 2. DÉFINIR LE MODÈLE

print("\n")
print("--Modèle de Régression linéaire--\n")

# Le modèle repose sur la fonction : y = w * x + b
# w = poids (pente de la ligne)
# b = biais (ordonnée à l'origine)

# Init ALÉATOIRE des paramètres
# (le modèle ne sait rien au début donc il va dabord deviner au hasard)
np.random.seed(42)  # Pour la reproduction
w = np.random.randn()  # Poids aléatoire
b = np.random.randn()  # Biais aléatoire

print("\nParamètres initiaux (aléatoires) :")
print(f"  w (poids) = {w:.4f}")
print(f"  b (biais) = {b:.4f}")

#Fonction de prédiction
def predict(x, w, b):
    # Rappel : y = w * x + b
    return w * x + b

#Test avec les paramètres de base
y_pred_initial = predict(X_train, w, b)

print("\n Prédictions avec paramètres aléatoires (AVANT l'apprentissage) : ")
for i in range(len(X_train)):
    print(f"  Pixels verts = {X_train[i]:2d}% --> Prédiction = {y_pred_initial[i]:6.2f}%, Réel = {y_train[i]:2d}%")

# Visualisation de la ligne initiale
plt.figure(figsize=(10, 6))
plt.scatter(X_train, y_train, color="green", s=100, alpha=0.6, label="Données réelles")
plt.plot(X_train, y_pred_initial, color="red", linewidth=2, label="Ligne initiale", linestyle="--")
plt.xlabel("Pourcentage de pixels verts (%)", fontsize=12)
plt.ylabel("Couverture forestière (%)", fontsize=12)
plt.title("Modèle AVANT apprentissage (avec les paramètres aléatoires)", fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

print("\nConclusion: La ligne ne coincide pas du tout avec les points !")
print("On doit changer w et b")

# 3. FONCTION D'ERREUR (LOSS FUNCTION)

print("--Fonction d'erreur (LOSS)--")

def compute_loss(X, y_true, w, b):
    """ Calcule l'erreur moyenne (Mean Squared Error(MSE)).
        Formule : Loss = (1/n) * Σ (prédiction - réalité)² 
        loss faible == meilleure prédiction."""
    y_pred = predict(X, w, b)
    errors = y_pred - y_true  # Différence entre  la prédiction et la réalité
    squared_errors = errors ** 2  # Elève au carré pour pénaliser les grandes erreurs
    mse = np.mean(squared_errors)  #Moyenne
    return mse

# Calcul de la loss
initial_loss = compute_loss(X_train, y_train, w, b)

print(f"\nLoss initiale avec les paramètres aléatoires : {initial_loss:.2f}")
print("\nResultats : En moyenne, les prédictions sont à +- {:.1f}% de la réalité".format(np.sqrt(initial_loss)))
print("Mauvais, à améliorer !")

# 4. APPRENTISSAGE (GRADIENT DESCENT)

print("\n")
print("--APPRENTISSAGE (GRADIENT DESCENT)--")

print("\n Ajuster w et b petit à petit pour réduire la loss")
print("On calcule dans quelle direction bouger w et b")
print("On fait un petit pas dans cette direction")
print("On répète jusqu'à ce que la loss soit minimale")

# Hyperparamètres
learning_rate = 0.0001  #Taille du pas (trop grand = instable, trop petit = lent)
n_iterations = 1000     #Nombre d'étapes d'apprentissage

print("\nHYPERPARAMÈTRES :")
print(f"   Learning rate : {learning_rate}")
print(f"   Itérations    : {n_iterations}")

# Historique
loss_history = []
w_history = []
b_history = []

# apprentissage
for iteration in range(n_iterations):
    # 1. On fait une prédiction
    y_pred = predict(X_train, w, b)
    
    # 2. In calcule l'erreur
    errors = y_pred - y_train
    
    # 3. On calcule les gradients (la direction pour ajuster w et b)
    # Gradient = dérivée de la loss par rapport au paramètres
    grad_w = (2 / len(X_train)) * np.sum(errors * X_train)
    grad_b = (2 / len(X_train)) * np.sum(errors)
    
    # 4. On met à jour  les paramètres (donc on fait un pas dans la direction qu'on croit bonne)
    w = w - learning_rate * grad_w  # Descente du gradient
    b = b - learning_rate * grad_b
    
    # 5. Et on calculer une nouvelle loss pour voir si le résultat est pls accurate
    loss = compute_loss(X_train, y_train, w, b)
    
    # Sauvegarde de  l'historique
    loss_history.append(loss)
    w_history.append(w)
    b_history.append(b)
    
    # Toutes les 100 itéractions on affiche la  progression
    if (iteration + 1) % 100 == 0:
        print(f"Itération {iteration+1:4d} | Loss = {loss:8.4f} | w = {w:7.4f} | b = {b:7.4f}")

print("\nOn a terminé la phase apprentissage.")

# 5. VISUALISATION DE L'APPRENTISSAGE

print("--VISUALISATION--")

# Graphique 1 : Évolution de la loss
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(loss_history, color="blue", linewidth=2)
plt.xlabel("Itérations", fontsize=12)
plt.ylabel("Loss (erreur)", fontsize=12)
plt.title("Évolution de l'erreur pendant l'apprentissage", fontsize=14)
plt.grid(True, alpha=0.3)

# Graphique 2 : Évolution des paramètres
plt.subplot(1, 2, 2)
plt.plot(w_history, label="w (poids)", color="red", linewidth=2)
plt.plot(b_history, label="b (biais)", color="orange", linewidth=2)
plt.xlabel("Itération", fontsize=12)
plt.ylabel("Valeur du paramètre", fontsize=12)
plt.title("Évolution des paramètres w et b", fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\nConclusion:")
print("La loss diminue progressivement aka le modèle devient meilleur")
print("Les paramètres w et b convergent vers des valeurs stables")

# 6. RÉSULTATS

print("--RÉSULTATS--")

final_loss = loss_history[-1]

print("\nParamètres finaux après apprentissage :")
print(f"  w = {w:.4f}")
print(f"  b = {b:.4f}")
print(f"\nLoss final : {final_loss:.4f}")
print(f"Erreur moyenne de : +-{np.sqrt(final_loss):.2f}%")

# Avant/après
print(f"\n Améliorations :")
print(f"   Loss initiale : {initial_loss:.2f}")
print(f"   Loss finale   : {final_loss:.2f}")
print(f"   Réduction     : {((initial_loss - final_loss) / initial_loss * 100):.1f}%")

# Prédictions finales
y_pred_final = predict(X_train, w, b)

print(f"\nPrédictions après apprentissage :")
for i in range(len(X_train)):
    error = abs(y_pred_final[i] - y_train[i])
    print(f"  Pixels verts = {X_train[i]:2d}% → Prédiction = {y_pred_final[i]:5.1f}%, Réel = {y_train[i]:2d}% (erreur = {error:.1f}%)")

# Visualisation
plt.figure(figsize=(10, 6))
plt.scatter(X_train, y_train, color="green", s=100, alpha=0.6, label="Données réelles", zorder=3)
plt.plot(X_train, y_pred_final, color="blue", linewidth=3, label="Ligne apprise", zorder=2)

# Afficher les erreurs (ici, lignes verticales)
for i in range(len(X_train)):
    plt.plot([X_train[i], X_train[i]], [y_train[i], y_pred_final[i]], 
             color="red", linestyle="--", alpha=0.5, linewidth=1, zorder=1)

plt.xlabel("Pourcentages de pixels verts (%)", fontsize=12)
plt.ylabel("Couverture forestière (%)", fontsize=12)
plt.title("Modèle après apprentissage (la meilleure ligne)", fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

#7. PREDICTIONS SUR DE NOUVELLES DONNÉES

print("🔮 PRÉDICTIONS SUR DE NOUVELLES DONNÉES")

# Nouvelles valeurs  que le modèle n'a jamais vu pendant l'entraînement
X_new = np.array([15, 35, 55, 75])

print("\nDonnées jamais vues par le modèle:")
y_new_pred = predict(X_new, w, b)

for i in range(len(X_new)):
    print(f"  Pixels verts = {X_new[i]:2d}% → Prédiction = {y_new_pred[i]:5.1f}%")
```
