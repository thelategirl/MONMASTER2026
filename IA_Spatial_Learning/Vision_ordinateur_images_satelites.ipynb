```python
# TEST n° 7 IMAGES SATELLITE
# Vision ordinateur

# 1. Import les bibliothèques 

import os  #navigation dossiers
import numpy as np  # calculs sur les arrays
import matplotlib.pyplot as plt  # afficher les images
from PIL import Image  # Charger les images

# Pour que les images s'affichent correctement dans Jupiter
%matplotlib inline

# 2.  CHARGER UNE IMAGE SATELLITE

dataset_path = "EuroSAT"
#category = "Forest"
category = "River"

# Construire un chemin complet vers le dossier de la catégorie
category_path = os.path.join(dataset_path, category)

# Lister tous les fichiers dans ce dossier
image_files = os.listdir(category_path)

# Prendre la première image de la liste
first_image_name = image_files[0]
image_path = os.path.join(category_path, first_image_name)

print(f"Image chargée : {image_path}")

# Charger l'image avec PIL
image = Image.open(image_path)

print("L'Image est chargée")


# 3.AFFICHER L'IMAGE

plt.figure(figsize=(6, 6))# On créer une image de 6x6 pouces
plt.imshow(image)
plt.title(f"Image satellite - Catégorie : {category}")
plt.axis('off') # Cacher les axes
plt.show()

print(f"Catégorie de l'image : {category}")


# 4. COMPRENDRE L'IMAGE
# Pour l'ordinateur, c'est un tableau de nombre, pas une forêt.

# Convertir l'image PIL en array
image_array = np.array(image)

print("\n--INFORMATIONS SUR L'IMAGE--")
print(f"Type de données : {type(image_array)}")
print(f"Dimensions (shape de l'array) : {image_array.shape}")
print(f"\nNombre total de pixels : {image_array.shape[0] * image_array.shape[1]}")
print(f"Nombre total de valeurs : {image_array.size}")  # hauteur × largeur × 3 canaux


# 5.COMPRENDRE LES VALEURS DES PIXELS

# On prends un pixel au centre de l'image
center_y = image_array.shape[0] // 2
center_x = image_array.shape[1] // 2
center_pixel = image_array[center_y, center_x]

print("\n--PIXEL DU CENTRE DE L'IMAGE--")
print(f"Position : ligne {center_y}, colonne {center_x}")
print(f"Valeurs RGB : {center_pixel}")
print(f"  → Rouge (R) : {center_pixel[0]}")
print(f"  → Vert (G) : {center_pixel[1]}")
print(f"  → Bleu (B) : {center_pixel[2]}")
print("\nChaque valeur va de 0 (noir) à 255 (maximum de cette couleur)")


# 6. STATS

print("\n--STATS")
red_channel = image_array[:, :, 0]    # Tous les pixels rouges
green_channel = image_array[:, :, 1]  # Tous les pixels verts
blue_channel = image_array[:, :, 2]   # Tous les pixels bleus

print("Canal rouge :")
print(f"  Valeur minimale : {red_channel.min()}")
print(f"  Valeur maximale : {red_channel.max()}")
print(f"  Valeur moyenne : {red_channel.mean():.2f}")

print("\nCanal vert :")
print(f"  Valeur minimale : {green_channel.min()}")
print(f"  Valeur maximale : {green_channel.max()}")
print(f"  Valeur moyenne : {green_channel.mean():.2f}")

print("\nCanal bleu:")
print(f"  Valeur minimale : {blue_channel.min()}")
print(f"  Valeur maximale : {blue_channel.max()}")
print(f"  Valeur moyenne : {blue_channel.mean():.2f}")

# Conslusion : Dans une image de forêt, le canal vert à des valeurs plus élevée

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(red_channel, cmap='Reds')
axes[0].set_title('Canal rouge (R)')
axes[0].axis('off')

axes[1].imshow(green_channel, cmap='Greens')
axes[1].set_title('Canal vert (G)')
axes[1].axis('off')

axes[2].imshow(blue_channel, cmap='Blues')
axes[2].set_title('Canal bleu (B)')
axes[2].axis('off')

plt.tight_layout()
plt.show()

print("Les zones CLAIRES = valeurs élevées de cette couleur")
print("Les zones SOMBRES = valeurs faibles de cette couleur")


# 8. Comparer plusieurs images de catégories différentes

categories_to_compare = ["Forest", "River", "Highway"]
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for i, cat in enumerate(categories_to_compare):
    cat_path = os.path.join(dataset_path, cat)
    
    if not os.path.exists(cat_path):
        print(f" La categorie {cat} n'existe pas")
        continue
    
    cat_files = os.listdir(cat_path)
    img_path = os.path.join(cat_path, cat_files[0])
    img = Image.open(img_path)
    
    axes[i].imshow(img)
    axes[i].set_title(f"{cat}")
    axes[i].axis('off')

plt.tight_layout()
plt.show()

print("\n--CONCLUSIONS--")
print("→ Les forêts sont vertes")
print("→ Les rivières ont du bleu")
print("→ Les routes sont grises/noires")
```
